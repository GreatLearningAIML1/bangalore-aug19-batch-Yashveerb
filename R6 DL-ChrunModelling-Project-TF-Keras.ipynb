```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df=pd.read_csv('C://Users//Sandhyash//Downloads//Churn_Modelling.csv')
df.info()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['Geography']=le.fit_transform(df['Geography'])
#pd.get_dummies(df,prefix=None, prefix_sep='',columns=['Gender', 'Geography'], drop_first=True)
df.head()
```

```python
df=df.drop(columns=['CustomerId', 'Surname', 'RowNumber'], axis=1)
df.head()
```

```python
#df=df.drop(columns=['CustomerId'], axis=1) already done in above step

X=df.drop(columns=['Exited']) # independent features
y=df['Exited'] #dependent or target feature
```

Normalize the train and test data (2.5 points)
 Initialize & build the model (10 points)
Optimize the model (5 points)
Predict the results using 0.5 as a threshold (5 points)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.02, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.head()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train= sc.fit_transform(X_train)
X_test= sc.fit_transform(X_test)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization
from sklearn.metrics import confusion_matrix
```

```python
model= Sequential()
b_norm= BatchNormalization()
```

```python
model.add(Dense(32, activation='relu',input_shape=(X_train.shape[1],)))
#model.add(Dense(activation='relu', input_dim=11, units=))
#model.add(Dense(5, activation='relu'))
```

```python
model.add(Dense(1, activation='sigmoid'))
```

```python
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='SGD', metrics=['accuracy'],loss='binary_crossentropy')
```

```python
model.summary()
```

```python
model.fit(X_train, y_train,
         epochs=50,
         validation_data=(X_test, y_test),
         batch_size=32)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
y_pred=y_pred>.5
y_pred
```

```python
cm= confusion_matrix(y_test, y_pred)
cm
```

Model clasification accuracy on validation data is 86.50%

Optimizing the model

```python
#model.add(b_norm)
model.add(Dense(32, activation='relu',input_shape=(X_train.shape[1],)))
#model.add(b_norm)
model.add(Dense(16, activation='relu'))
#model.add(b_norm)
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', metrics=['accuracy'],loss='binary_crossentropy')
```
